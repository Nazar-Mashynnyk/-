```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math as m

data=pd.read_csv('d.csv',delimiter='\t')['Вар.№4']
ranges=np.linspace(data.min(),data.max(),1+int(m.log2(len(data))))
interval_row=pd.cut(data,bins=ranges, right=False).value_counts().sort_index()
print('Інтервальний статистичний ряд\nДіапазон  :   Кількість значень всередині діапазону')
for row in interval_row.index:
    print(f'{row}   :   {interval_row[row]}')
```

```python
plt.plot(sorted(data), np.linspace(0, 1, len(data), endpoint=False))
plt.title('Полігон')
plt.grid(True)
plt.show()
```

```python
plt.hist(data, bins=1+int(m.log2(len(data))), density=True, alpha=0.6, color='g')
plt.title('Гістограма')
plt.grid(True)
plt.show()
```

```python
x = np.sort(data)
y = np.arange(1, len(x) + 1) / len(x)
plt.step(x, y)
plt.title('Емпірична функція розподілу')
plt.grid(True)
plt.show()
```

```python

plt.plot(sorted(data), np.linspace(0, 1, len(data), endpoint=False).cumsum())
plt.axvline(np.median(data), color='red', linestyle='--', label='Медіана')
plt.axvline(np.percentile(data, 25), color='green', linestyle='--', label='25-й квартиль')
plt.axvline(np.percentile(data, 75), color='purple', linestyle='--', label='75-й квартиль')
plt.title('Кумулятивна крива')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f"Вибіркове середнє: {np.mean(data)}")
print(f"Дисперсія: {np.var(data)}")
print(f"Середньоквадратичне відхилення: {np.std(data)}")
print(f"Мода: {data.value_counts().index[0]}")
print(f"Медіана: {np.median(data)}")
print(f"Коефіцієнт асиметрії: {data.skew()}")
print(f"Коефіцієнт ексцесу: {data.kurt()}")
```
